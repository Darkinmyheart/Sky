```python
from tkinter import *

window = Tk()
#window.resizalbe(False,False)

#window.geometry("600x150")

entry = Entry(window, borderwidth = 4)
entry.grid(row=0,column = 2,padx = 5,pady = 10)
def covert ():
    weight = float(entry.get())
    gram = weight * 1000
    pounds = weight * 2.20462
    ounce = weight * 35.274
    Label_rsgram["text"] = gram
    Label_rsounce["text"]= ounce
    Label_rspounds["text"]= pounds


Label_enter = Label(window,text=("Enter the weight in kg"))
Label_gram = Label(window,text=("Gram"))
Label_pounds = Label(window,text=("Pounds"))
Label_ounce = Label(window,text = ("Ounce"))


Label_rsgram = Label(window,text = "")
Label_rspounds = Label(window,text="")
Label_rsounce = Label(window,text = "")


Label_rsgram.grid(row=2,column = 0)
Label_rsounce.grid(row=2,column = 5)
Label_rspounds.grid(row=2,column = 2)
Label_enter.grid(row=0,column = 0)
Label_gram.grid(row=1,column = 0)
Label_pounds.grid(row=1,column = 2)
Label_ounce.grid(row=1,column = 5)

Button_convert = Button(window, text = ("comvert"),font= ("times",10),command = covert)
Button_convert.grid(row = 0, column = 4)

window.mainloop()
```
