```python
import pandas as pd
df =pd.read_csv("subset-covid-data.csv")
df.head()
```

```python
df.describe(include='all')
pd.unique(df.continent)
```

```python
df.info()
```

```python
df.date.value_counts()
```

```python
cleaned_df = df[df.date == '2020-04-12']
print(f"Trung bình các ca mắc mới : {str(cleaned_df.cases.mean())}")
print ("trung vị của số ca mắc mới: "+ str(cleaned_df.cases.median()))
import matplotlib.pyplot as plt
plt.hist(cleaned_df.cases, bins = 200)
plt.title("Phân bố số ca mắc mới")
plt.xlabel("số số ca mắc mới")
plt.ylabel("Số lượng quốc gia")
```

```python
cleaned_df.groupby('continent')['deaths','cases'].sum()
```

```python
cleaned_df.groupby('continent')['deaths','cases'].mean()
```
