```python
so1 = int(input("Nhập số bắt đầu :"))
so2 = int(input("Nhập số kết thúc :"))
if so1 < so2 :
    for i in range (so1,so2):
        if (i %3 == 0 and i%5==0) :
            print ('FizzBuzz',end = ' ')
        elif (i%3 == 0 ):
            print('Fizz',end = ' ')
        elif (i%5 == 0) :
            print ('Buzz',end = ' ')
        else:
            print (i, end = ' ')
else :
    print ('Nhập lại số bắt đầu và số kết thúc :')
```
