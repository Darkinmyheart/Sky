### Thực hành: Viết chương trình tạo ra một list các string khác nhau và đếm tổng số lượng string dài hơn 3 kí tự

```python
list_string = ['âdfafsdfdfsff', 'faffdfffwf', "sgfffgg", 'a']
def dem_string():
    num = 0
    for i in list_string:
        if len(i)>3:
            num += 1
    print(num)
dem_string()
```

### Kiểm tra phần tử tồn tại trong list

```python
a = [10,20,30,40,50]
# Loop
for item in a :
    if item == 30 :
        print ("30 co ton tai trong a")
        
# in
if 30 in a:
    print("30 co ton tai trong a")
else :
    pass
# count
if a.count(30) >0:
    print("30 co ton tai trong a")
else:
    pass
```

### Sắp xếp list

```python
a = [10,20,30,40,50]
# sort sẽ thực hiện sắp xếp lại list và làm thay đổi trực tiếp thứ tự của các element
a.sort(reverse = True)   # true là chiều từ lớn đến bé , false là ngược lại (mặc định là false)
print(a)
# sorted sẽ thực hiện sắp xếp lại list nhưng không làm thay đổi trực tiếp thứ tự của các element
# mà tạo ra một list mới và return về
b = sorted(a,reverse = True)
print (b)
# sắp xếp chữ theo nguyên tắc bảng chữ cái và chữ hoa trước chữ thường
```
