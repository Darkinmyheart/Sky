### Xuất dữ liệu ra màn hình

### Các cách custom với lệnh print

Để in một nội dung kèm theo giá trị của biến, ta có các cách sau:

```python
# Ilove chocolate and I hate sandwich
food_1 = "chocolate"
food_2 = "sandwich"
```

```python
# In trực tiếp
print(" I love chocolate and I hate sandwich ")
```

```python
# In bằng cách ghép các nội dung với nhau
food_1 = "chocolate"
food_2 = "sandwich"
print(" I love ", food_1, "and I hate ", food_2)
# để thay đổi ký tự nối giữa các nội dung trong print, chúng ta sẽ sử dụng sep=''
# sep viết tắt của separate
print("I love", food_1, "and", food_2, sep='+') # casting convention
```

```python
# Sử dụng String Format
print(f"I love {food_1} and I love {food_2}")
```

```python
# Sử dụng nhiều lệnh print
print("I love ", end="")
print(food_1)
print("and")
print("I love")
print(food_2)
```

#### Thực hành: Viết chương trình nhập vào chiều dài và chiều rộng của hình chữ nhật, sau đó in ra chu vi và diện tích của hình chữ nhật đó

```python
# Nhập các kích thước của hình chữ nhật
a = int(input(" Hãy nhập chiều dài của hình chữ nhật : "))

print (a)
```

```python
# Tính toán chu vi, diện tích
chieu_dai = int(input("Nhập chiều dài hình chữ nhật :"))
chieu_rong = int(input(" Nhập chiều rộng hình chữ nhật :"))
chu_vi = (chieu_dai + chieu_rong) * 2
dien_tich = chieu_dai * chieu_rong
print (" Chu vi hình chữ nhật bằng :", chu_vi)
print (" Diện tích hình chữ nhật :", dien_tich)
```

```python
# In kết quả ra màn hình bằng 3 cách
print (f"Chu vi hình chữ nhật là {chu_vi} và diện tích hình chữ nhật là {dien_tich}")
print ("Chu vi hình chữ nhật là :",end=" ")
print ( chu_vi)
print ("Diện tích hình chữ nhật là :",end=" ")
print (dien_tich)
```
