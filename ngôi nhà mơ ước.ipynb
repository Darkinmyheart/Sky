```python
import math
import turtle
pen = turtle.Turtle()
pen.speed(10)

turtle.bgcolor("#05dbf1")
def move_pen (x,y) :
    u=-100
    pen.penup()
    pen.goto(x,y)
    pen.pendown()
    
#vẽ tường chính
move_pen(-345,-200)
pen.fillcolor("#3957f2")
pen.begin_fill()
for i in range(1,3):
    pen.fd(150)
    pen.lt(90)
    pen.fd(200)
    pen.lt(90)
pen.end_fill()

#ve cửa
move_pen(-300,-200)
pen.fillcolor ("#4ff559")
pen.begin_fill()
for i in range(1,3):
    pen.fd(60)
    pen.lt(90)
    pen.fd(110)
    pen.lt(90)
pen.end_fill()

#Vẽ tường
move_pen(-195,-200)
pen.fillcolor("#f5f362")
pen.begin_fill()
pen.lt(25)
pen.fd(180)
pen.lt(65)
pen.fd(200)
pen.lt(115)
pen.fd(180)
pen.lt(65)
pen.fd(200)
pen.lt(90)
pen.end_fill()

#vẽ cửa sổ
move_pen(-140,-100)
pen.fillcolor("#ac6517")
pen.begin_fill()
pen.lt(25)
pen.fd(60)
pen.lt(65)
pen.fd(60)
pen.lt(115)
pen.fd(60)
pen.lt(65)
pen.fd(60)
pen.lt(90)
pen.end_fill()

#Vẽ mái nhỏ
move_pen(-195,0)
pen.fillcolor("#ed23e8")
pen.begin_fill()
pen.lt(135)
pen.fd(math.sqrt(1/2)*150)
pen.lt(90)
pen.fd(math.sqrt(1/2)*150)
pen.lt(135)
pen.fd(150)
pen.end_fill()

#Vẽ mái nhà lớn
pen.fillcolor("#fbc950")
pen.begin_fill()
pen.lt(25)
pen.fd(180)
pen.lt(110)
pen.fd(math.sqrt(1/2)*150)
pen.lt(70)
pen.fd(180)
pen.lt(110)
pen.fd(math.sqrt(1/2)*150)
pen.lt(45)
pen.end_fill()

#Vẽ cái cây
move_pen(200,-250)
pen.fillcolor("#a97f1a")
pen.begin_fill()
for i in range (1,3):
    pen.fd(80)
    pen.lt(90)
    pen.fd(150)
    pen.lt(90)
pen.end_fill()

move_pen(240,-100)
pen.fillcolor("#1deb54")
pen.begin_fill()
for i in range (1,4):
    pen.fd(100)
    pen.lt(135)
    pen.fd(math.sqrt(1/2)*200)
    pen.lt(90)
    pen.fd(math.sqrt(1/2)*200)
    pen.lt(135)
    pen.fd(100)
    pen.penup()
    pen.lt(90)
    pen.fd(100)
    pen.rt(90)
    pen.pendown()
pen.end_fill()

move_pen(550,250)
pen.fillcolor("yellow")
pen.begin_fill()
for i in range (1,5):
    pen.circle(50,extent=45)
    pen.rt(90)
    pen.fd(20)
    pen.lt(180)
    pen.fd(20)
    pen.rt(90)
    pen.circle(50,extent=45)
    pen.rt(90)
    pen.fd(50)
    pen.lt(180)
    pen.fd(50)
    pen.rt(90)

pen.end_fill()

turtle.done()
```
